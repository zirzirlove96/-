```python
parent=dict()
rank=dict()

def merge_sort(array):
    if len(array)<=1:
        return array
    mid=len(array)//2

    array1=merge_sort(array[:mid])
    array2=merge_sort(array[mid:])

    result=[]
    while array1 and array2:
        if array1[0][0]<array2[0][0]:
            result.append(array1.pop(0))
        else:
            result.append(array2.pop(0))
    while array1:
        result.append(array1.pop(0))
    while array2:
        result.append(array2.pop(0))

    return result

def union(node_v,node_u):
    root1=find(node_v)
    root2=find(node_u)

    if rank[root1]>rank[root2]:
        parent[root2]=root1
    else:
        parent[root1]=root2
        if rank[root1]==rank[root2]:
            rank[root1]+=1

def find(node):
    if parent[node]!=node:
        parent[node]=find(parent[node])
    return parent[node]

def make_set(node):
    parent[node]=node
    rank[node]=0


def kruskal(graph):

    mst = list()

    for node in graph['vertices']:
        make_set(node)

    edges=graph['edges']
    edges=merge_sort(edges)


    for edge in edges:
        weight,node_v,node_u=edge
        if find(node_v)!=find(node_u):
            union(node_v,node_u)
            mst.append(edge)

    return mst

n=int(input())

array=[]
for _ in range(n):
    x,y,z=map(int, input().split())
    array.append([x,y,z])

graph={
    'vertices':[],
    'edges':[]
    }

for i in range(1,n+1):
    graph['vertices'].append(i)

for i in range(len(array)-1):
    max_m=10**9+1
    for j in range(i+1,len(array)):
        x=array[i][0]-array[j][0]
        y=array[i][1]-array[j][1]
        z=array[i][2]-array[j][2]
        value=min(abs(x),abs(y),abs(z))
        if value<max_m:
            max_m=value
            graph['edges'].append([value,i+1,i+2])

graph['edges']=merge_sort(graph['edges'])
#print(graph)

s=0
for a in kruskal(graph):
    #print(a)
    s+=a[0]

print(s)
    
```

```python
parent=dict()
rank=dict()


def union(node_v,node_u):
    root1=find(node_v)
    root2=find(node_u)

    if rank[root1]>rank[root2]:
        parent[root2]=root1
    else:
        parent[root1]=root2
        if rank[root1]==rank[root2]:
            rank[root1]+=1

def find(node):
    if parent[node]!=node:
        parent[node]=find(parent[node])
    return parent[node]

def make_set(node):
    parent[node]=node
    rank[node]=0


def kruskal(graph):

    mst = list()

    for node in graph['vertices']:
        make_set(node)

    edges=graph['edges']
    edges.sort()


    for edge in edges:
        weight,node_v,node_u=edge
        if find(node_v)!=find(node_u):
            union(node_v,node_u)
            mst.append(edge)

    return mst

n=int(input())

graph={
    'vertices':[],
    'edges':[]
    }

planet=[]
for i in range(1,n+1):
    graph['vertices'].append(i)

for i in range(1,n+1):
    x,y,z=map(int, input().split())
    planet.append([x,y,z,i])
    
for i in range(3):
    planet.sort(key=lambda x:x[i])
    before=planet[0][3]
    for j in range(1,n):
        #가까운 행성 즉 1,2
        #처럼 가까운 행성들의 거리를 graph에 넣어준다.
        current=planet[j][3]
        graph['edges'].append([abs(planet[j][i]-planet[j-1][i]),before,current])
        before=current
        #before과 current를 해줘서 간선의 갯수를 줄여 준다.
        #quick_sort, sys를 사용해줘도 시간초과가 나는 것을 보아 여기에서 초과가 나지 않도록 해야 한다.

print(graph)
s=0
for a in kruskal(graph):
    #print(a)
    s+=a[0]

print(s)
    
```
