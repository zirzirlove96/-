# Heap 자료구조

## heapq에 대해 알아보자.

<pre>
힙은 특정한 규칙을 가지는 트리로, 최댓값과 최솟값을 찾는 연산을 빠르게 하기 위해 고안된 완전이진트리를 기본으로 한다.

최소 힙 : 부모 노드의 키 값이 자식 노드의 키 값보다 작은 힙
최대 힙 : 부모 노드의 키 값이 자식 노드의 키 값보다 항상 큰 힙

파이썬 heapq 모듈을 heapq(priority queue)알고리즘을 제공한다.
<b>내부적으로 인덱스 0에서 시작해서 k번째 원소가 항상 자식 원소들(2k+1, 2k+2)보다 작거나 같은 최소 힙의 형태로 정렬된다.</b>
</pre>

<pre>
heapq.heappush(heap,item):item을 heap에 추가
heapq.heappop(heap): heap에서 제일 작은 원소를 pop 비어있을 경우 indexError
heapq.heapify(x): 리스트 x를 heap으로 변환함
</pre>

```python
#최소 힙
import heapq

heap = []
heapq.heappush(heap, 50)
heapq.heappush(heap, 10)
heapq.heappush(heap, 20)
heapq.heappush(heap, 60)
heapq.heappush(heap, 15)
heapq.heappush(heap, 30)
heapq.heappush(heap, 25)

print(heap)
#부모 노드들만을 자식노드들보다 작거나 같은 값으로 정렬해 준다,
```

```python
# 최대 힙 만들기
import heapq
heap_items = [1,3,5,7,9]

max_heap = []
for item in heap_items:
  heapq.heappush(max_heap, (-item, item))#튜플로 첫 번째 원소를 우선순위로 구성된다.

print(max_heap)
```
